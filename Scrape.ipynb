```python
# set URL
url = 'https://riwayat-file-vaksinasi-dki-jakarta-jakartagis.hub.arcgis.com'
driver.get(url)
time.sleep(5)

# Execute
for i in range(1,106):
    print(f'Iter {i}')
    
    # Click link to Google Sheets
    xpath = f'//*[@id="ember86"]/table/tbody/tr[2]/td[1]/p[{i}]/a'
    button = driver.find_element_by_xpath(xpath)
    button.click()
    print(f'  Succesfully clicked link -> Current URL: {driver.current_url}')

    # Wait for gsheets tab to load
    time.sleep(2)

    # Switch to gsheets tab
    drive_tab = driver.window_handles[-1]
    driver.switch_to.window(drive_tab)
    driver.switch_to_window(drive_tab)
    print(f'  Switched tab to: {driver.current_url}')

    # Press button to automatically download 
    # (only if has switched to gsheets tab, as confirmed by 'drive' in current URL)
    if 'drive' in driver.current_url:
        download = driver.find_element_by_xpath('/html/body/div[3]/div[4]/div/div[3]/div[2]/div[2]/div[3]/div')
        download.click()
        print(f'  Downloaded file.')

        time.sleep(2)

        # Close tab twice (gsheet and downloads tab)
        driver.close()
#         driver.close()

        # Revert to original tab
        orig_tab = driver.window_handles[0]
        driver.switch_to.window(orig_tab)
        driver.switch_to_window(orig_tab)
        print(f'  Switched to original tab: {driver.current_url}')
    
    print('\n')
```
